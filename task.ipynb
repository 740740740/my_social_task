```python
import networkx as nx
```

```python
s=nx.Graph()
node_list=(1,2,3,4,5,6)
s.add_nodes_from(node_list)
edge_list=((1,4),(1,2),(1,3),(1,6),(2,3),(2,4),(3,6))
s.add_edges_from(edge_list)
nx.draw(s,with_labels=True)
```

```python
Adj = nx.adjacency_matrix(s).todense()
Adj
```

```python
from collections import Counter
import matplotlib.pyplot as plt

degree_of_sequence = [s.degree(d) for d in s.nodes]
degree_counts = Counter(degree_of_sequence)
min_degree, max_degree = min(degree_counts.keys()), max(degree_counts.keys())
plt.xlabel("Degree", fontsize=20)
plt.ylabel("Number of Nodes", fontsize=20)
plot_x = list(range(min_degree, max_degree + 1))
plot_y = [degree_counts.get(x, 0) for x in plot_x]
plt.bar(plot_x, plot_y)
```

```python
def plot_degree_dist(s):
    
    degrees = s.degree()
    degrees = dict(degrees)
    values = sorted(set(degrees.values()))
    print(values)
    histo = [list(degrees.values()).count(x) for x in values]
    P_k = [x / s.order() for x in histo]
    print(len(P_k))
    
    plt.figure()
    plt.bar(values, P_k)
    plt.xlabel("k",fontsize=40)
    plt.ylabel("p(k)", fontsize=40)
    plt.title("Degree Distribution", fontsize=40)
    
    plt.show()
    plt.figure()
    plt.grid(False)
    plt.loglog(values, P_k, "bo")
    plt.xlabel("k", fontsize=40)
    plt.ylabel("log p(k)", fontsize=40)
    plt.title("log Degree Distribution")
    plt.show()
    plt.show() 
```

```python
plot_degree_dist(s)
```
